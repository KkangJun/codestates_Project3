```python
video1 = 'E:\\vehicle_front_camera_20200904_gangnam\drive\\1.MP4'
video2 = 'E:\\vehicle_front_camera_20200904_gangnam\drive\\2.MP4'

cap1 = cv2.VideoCapture(video1)
# cap2 = cv2.VideoCapture(video2)

if cap1.isOpened():
  while True:
    ret, img = cap1.read()
    
    if ret:
      
      results = model(img)
      preds = results.pred[0].cpu().numpy()
      for pred in preds:
        x1, y1, x2, y2, acc, cl = pred
        label = class_list[cl]
        accu = np.round(acc, 2)
        
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        
        cv2.rectangle(img, (x1, y1), (x2, y2), (0,0,255), 5)
        cv2.putText(img, class_list[cl] + '\t' + str(accu), (x1, y1-20), cv2.FONT_ITALIC, 0.5, (0,255,0), 1)
      
      cv2.imshow('yolo_test', img)
      key = cv2.waitKey(33)
      
      if key == ord('q'):
        break
    else:
      break
else:
  print("can't open video.")

cap1.release()
# cap2.release()
cv2.destroyAllWindows()
```
